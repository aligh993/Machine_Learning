```python
# @author: ALI GHANBARI
# @email: alighanbari446@gmail.com

import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets, linear_model

m = 6.0;   b = 3.0;   noise = 0.3;

InlierX = np.random.rand(1000)
InlierY = m*InlierX+b+noise*np.random.randn(InlierX.shape[0])
data = np.column_stack([InlierX, InlierY])

m2 = 0.6;   noise2 = 0.4;
m3 = 0.1;   b3 = 4.5;   noise3 = 0.3;

OutlierX = m2+noise2*np.random.rand(100)
OutlierY = m3*OutlierX+b3+noise3*np.random.randn(OutlierX.shape[0])


linear = linear_model.LinearRegression()
linear.fit(data[:, 0].reshape(-1,1), data[:, 1].reshape(-1,1))


lineX = np.arange(data[:, 0].min(), 2)[:, np.newaxis]
lineY_linear = linear.predict(lineX)


# plt.plot(InlierX, InlierY, 'b.')
plt.plot(data[:, 0], data[:, 1], '.')
plt.plot(OutlierX, OutlierY, 'r.')
plt.plot(lineX, lineY_linear, color='g', linewidth=2, label='Linear regressor')
plt.show()
```
