```python
a = np.random.rand(1000)
b = np.random.rand(1000) + 1.1
x=np.concatenate((a, a, b, b))
y=np.concatenate((a, b, a, b))
df=pd.DataFrame({'x': x, 'y': y})
plt.scatter(x, y)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
plt.scatter(df.x, df.y,c=kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
plt.scatter(df.x, df.y,c=kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df)
plt.scatter(df.x, df.y,c=kmeans.labels_)
plt.show()
```

```python
elbow=[]
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    elbow.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), elbow)
plt.show()
```

```python
x = [1, 1, 1, 1, 1, 1, 2, 2, 2, 2]
y = [110000, 111000, 112000, 200000, 202000, 198000, 110000, 111000, 112000, 111500]
features = pd.DataFrame({'X' : x, 'Y': y})
features.X = (features.X - features.X.min()) / (features.X.max() - features.X.min()) 
features.Y = (features.Y - features.Y.min()) / (features.Y.max() - features.Y.min()) 
plt.scatter(features.X, features.Y)
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.show()
```

```python
x = [1, 1, 1, 1, 1, 1, 2, 2, 2, 2]
y = [110000, 111000, 112000, 200000, 202000, 198000, 110000, 111000, 112000, 111500]
features = pd.DataFrame({'X' : x, 'Y': y})
features.X = (features.X - features.X.mean()) / (features.X.std()) 
features.Y = (features.Y - features.Y.mean()) / (features.Y.std()) 
plt.scatter(features.X, features.Y)
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.show()
```
